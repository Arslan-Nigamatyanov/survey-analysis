```python
# Starter code for using visualization techniques:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('../data/cleaned_combined_survey.csv', sep=';')

print(f'Dataset shape: {df.shape}')
print(df.head(10))
```

```python
# === VISUALIZATION TECHNIQUES ===

# 1. Histogram - Distribution of a Single Variable

# Histogram of anxiety scores
plt.figure(figsize=(10, 6))
df['anxiety_score'].hist(bins=9, color='skyblue', edgecolor='black')
plt.title('Distribution of Anxiety Scores')
plt.xlabel('Anxiety score (1-10)')
plt.ylabel('Number of Patients')
plt.grid(False)
plt.savefig('../plots/anxiety_histogram.png', dpi=300, bbox_inches='tight')
plt.show
```

```python
#. Bar Plot - Compare Categories

# Average anxiety by age group
plt.figure(figsize=(10, 6))
age_group_means = df.groupby('age_group')['anxiety_score'].mean()

age_group_means.plot(kind='bar', color=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Average Anxiety Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Anxiety Score')
plt.xticks(rotation=45)
plt.savefig('../plots/anxiety_by_age_group.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 3. Box Plot - Distribution by Groups

# Box plot: anxiety distribution by age group
plt.figure(figsize=(10, 6))
df.boxplot(column='anxiety_score', by='age_group', grid=False)
plt.title('Anxiety Score Distribution by Age Group')
plt.suptitle('') # Remove automatic title
plt.xlabel('Age Group')
plt.ylabel('Anxiety Score')
plt.savefig('../plots/anxiety_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
